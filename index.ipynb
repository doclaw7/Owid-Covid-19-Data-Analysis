```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

data = pd.read_csv('owid-covid-data.csv')
```

```python
df = data

# checking the dataset information
print("The dataset columns are:")
display(df.columns)

print("\nThe dataset Rows are: ")
display(df.head())

#identifying the missing values
print("\nThis are the total missing values: ")
display(df.isnull().sum())
```

```python
# Drop rows where any of the specified columns have missing values
df_cleaned = df.dropna(subset=['date', 'location', 'total_cases'])

# Preview the cleaned data
display(df_cleaned)
```

```python
# Filter for the three countries
countries = ['Kenya', 'United States', 'India']
filtered_df = df[df['location'].isin(countries)]


# Drop rows with missing data
filtered_df = filtered_df.dropna(subset=['date', 'total_cases'])

# Convert 'date' to datetime
filtered_df['date'] = pd.to_datetime(filtered_df['date'])

# Pivot to get countries as columns
pivot_df = filtered_df.pivot(index='date', columns='location', values='total_cases')

# Plot
plt.figure(figsize=(14, 6))
for country in countries:
    plt.plot(pivot_df.index, pivot_df[country], label=country)

plt.title('Total COVID-19 Cases Over Time: Kenya vs India vs USA')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Bar Chart - Average alcohol content by quality
plt.figure(figsize=(14, 6))

df_grouped = filtered_df.groupby('location')['total_cases'].sum()


df_grouped.plot(kind='bar', color='skyblue')
plt.title('Countries vs Total cases')
plt.xlabel('Countries')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()
```

```python
# cummulative vaccinations over time for selected countries
# Filter data for cumulative vaccinations
vaccination_columns = ['date', 'location', 'total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated']
vaccination_data = filtered_df[vaccination_columns].dropna()

# Pivot data for cumulative vaccinations
vaccination_pivot = vaccination_data.pivot(index='date', columns='location', values='total_vaccinations')

# Plot cumulative vaccinations over time
plt.figure(figsize=(14, 6))
for country in countries:
    plt.plot(vaccination_pivot.index, vaccination_pivot[country], label=country)

plt.title('Cumulative COVID-19 Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculating % vaccinated population
population_data = filtered_df[['location', 'population']].drop_duplicates()
vaccinated_population = vaccination_data.groupby('location')['people_vaccinated'].max()
population_data.set_index('location', inplace=True)
vaccinated_percentage = (vaccinated_population / population_data['population']) * 100

# Plot pie charts for vaccinated vs. unvaccinated
for country in countries:
    vaccinated = vaccinated_percentage[country]
    unvaccinated = 100 - vaccinated
    plt.figure(figsize=(6, 6))
    plt.pie([vaccinated, unvaccinated], labels=['Vaccinated', 'Unvaccinated'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
    plt.title(f'Vaccinated vs. Unvaccinated Population in {country}')
    plt.show()
```

```python
import plotly.express as px

# Prepare the dataframe with iso_code and total_cases for the latest date
latest_data = df_cleaned[df_cleaned['date'] == df_cleaned['date'].max()]
choropleth_data = latest_data[['iso_code', 'total_cases']].dropna()

# Plot the choropleth map
fig = px.choropleth(
    choropleth_data,
    locations="iso_code",
    color="total_cases",
    hover_name="iso_code",
    title="COVID-19 Total Cases by Country",
    color_continuous_scale=px.colors.sequential.Plasma
)

fig.update_geos(showcoastlines=True, coastlinecolor="LightGray")
fig.show()
```
